```python
categorical_vars = ['district', 'urban', 'gender', 'highest_grade_completed',
                    'mm_account_telco', 'mm_account_telco_main']

df[categorical_vars] = df[categorical_vars].astype('category')
```

```python
# Create a dictionary to map old column names to new column names
rename_dict = {
    'start_time': 'survey_start_time',
    'end_time': 'survey_end_time',
    'hhid': 'household_id',
    'account_num': 'number_of_account',
    'weight': 'sampling_weight',
    'urban': 'urban_or_rural',
    'hh_members': 'household_members',
    'highest_grade_completed': 'highest_education_completed',
    'mm_account_cancelled': 'mobile_money_account_cancelled',
    'prefer_cash': 'preference_for_cash',
    'mm_trust': 'trust_in_mobile_money',
    'mm_account_telco': 'mobile_money_providers',
    'mm_account_telco_main': 'main_mobile_money_account',
    'v234': 'understood_terms_conditions',
    'agent_trust': 'trust_in_mobile_money_agents',
    'v236': 'taken_mobile_money_loan',
    'v237': 'network_unavailable_issues',
    'v238': 'clear_about_transaction_fees',
    'v240': 'failed_mobile_money_transaction',
    'v241': 'agent_without_enough_funds',
    'v242': 'has_copy_of_terms_conditions',
    'v243': 'know_where_to_complain',
    'v244': 'successfully_resolved_complaint',
    'v245': 'data_collection_by_providers',
    'v246': 'victim_of_fraud',
    'educational_level': 'educational_level',
    'financial_exclusion': 'financially_excluded',
    'digital_financial_inclusion': 'digital_financially_included'
}

# Rename the columns using the dictionary
df.rename(columns=rename_dict, inplace=True)

# Display the updated column names
print(df.columns)
```

```python
df.columns
```

```python
# Define a custom function to map existing educational levels to new categories
def categorize_educational_level(grade_completed):
    if 'primary' in grade_completed.lower():
        return 'primary'
    elif 'secondary' in grade_completed.lower():
        return 'secondary'
    elif 'tvet' in grade_completed.lower():
        return 'TVET'
    elif 'university' in grade_completed.lower():
        return 'university'
    else:
        return 'others'

# Create the new column 'educational_level' by applying the custom function
df['educational_level'] = df['highest_education_completed'].apply(categorize_educational_level)
```

```python
def is_digital_financially_included(row):
    account_type = row['account_type']
    return (account_type in ['Mobile Money', 'Online Bank Account'])

# Create the dummy variable for digital financial inclusion based on the criteria
df['digital_financially_included'] = df.apply(is_digital_financially_included, axis=1).astype(int)

# Calculate overall rate of digital financial inclusion
overall_digital_financial_inclusion_rate = df['digital_financially_included'].mean() * 100

print(df['digital_financially_included'].value_counts())
print("Overall Rate of Digital Financial Inclusion:", round(overall_digital_financial_inclusion_rate, 2))
```

```python
# Define functions to check financial exclusion and digital financial inclusion
def is_financially_excluded(row):
    return row['account_type'] == 'None' or row['account_type'] is None

df['financially_excluded'] = df.apply(is_financially_excluded, axis=1).astype(int)

overall_financial_exclusion_rate = df['financially_excluded'].mean() * 100

print(df['financially_excluded'].value_counts())
print("Overall Rate of Financial Exclusion:", round(overall_financial_exclusion_rate, 2))
```

# Plotting

```python
# Financial Inclusion Analysis by Age Group
age_groups = pd.cut(df['age'], bins=[18, 30, 40, 50, 60, 100])
financial_exclusion_by_age = df.groupby(age_groups)['financially_excluded'].mean() * 100
plt.figure(figsize=(10, 6))
financial_exclusion_by_age.plot(kind='bar')
plt.title('Financial Exclusion Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```

```python
# Function to Plot Distribution
def plot_distribution(data, title, xlabel, ylabel, bins=10):
    plt.figure(figsize=(10, 6))
    sns.histplot(data, bins=bins, kde=True)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

# Function to Plot Percentage Distribution
def plot_percentage_distribution(data, title, xlabel, ylabel):
    counts = data.value_counts(normalize=True) * 100
    plt.figure(figsize=(8, 6))
    counts.plot(kind='bar')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=0)
    plt.show()
```

```python
# Age Distribution Analysis
plot_distribution(df['age'], 'Age Distribution', 'Age', 'Frequency', bins=20)
```

```python
# Gender Analysis
plot_percentage_distribution(df['gender'], 'Gender Distribution', 'Gender', 'Percentage')
```

```python
# Education Level Analysis
plot_percentage_distribution(df['educational_level'], 'Education Level Distribution', 'Education Level', 'Percentage')
```

```python
# Location Analysis
plot_percentage_distribution(df['district'], 'Location Distribution', 'District', 'Percentage')
```

```python
# Household Size Analysis
plot_distribution(df['household_members'], 'Household Size Distribution', 'Household Size', 'Frequency', bins=10)
```

```python
def plot_categorical_group(dataframe, x_col, y_col, plot_title, x_label, y_label, rotation=45):
    group_data = dataframe.groupby(x_col)[y_col].mean()
    
    plt.figure(figsize=(10, 6))
    group_data.plot(kind='bar')
    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.xticks(rotation=rotation)
    plt.show()
```

```python
# Call the function with your DataFrame and specific columns
plot_categorical_group(df, 'educational_level', 'number_of_account', 'Average Number of Accounts by Education Level', 'Education Level', 'Average Number of Accounts')
```

```python
def plot_categorical_distribution(dataframe, column, plot_title, x_label, rotation=0):
    category_counts = dataframe[column].value_counts(normalize=True) * 100
    
    plt.figure(figsize=(10, 6))
    category_counts.plot(kind='bar')
    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel('Percentage')
    plt.xticks(rotation=rotation)
    plt.show()
```

```python
# Call the function to plot the distribution of main mobile money accounts
plot_categorical_distribution(df, 'main_mobile_money_account', 'Distribution of Main Mobile Money Accounts', 'Main Mobile Money Account', rotation=0)
```

```python
# Call the function to plot the distribution of mobile money providers
plot_categorical_distribution(df, 'mobile_money_providers', 'Distribution of Mobile Money Providers', 'Mobile Money Provider', rotation=45)
```

```python
def plot_group_by(dataframe, group_col, value_col, plot_title, x_label, y_label, rotation=0, bins=None):
    grouped_data = dataframe.groupby(group_col)[value_col].mean() * 100
    
    plt.figure(figsize=(10, 6))
    grouped_data.plot(kind='bar')
    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.xticks(rotation=rotation)
    plt.show()
```

```python
# Call the function with different columns
plot_group_by(df, 'educational_level', 'financially_excluded', 'Financial Exclusion Rate by Education Level', 'Education Level', 'Percentage')
```

```python
# Call the function with different columns and custom rotation
plot_group_by(df, 'urban_or_rural', 'financially_excluded', 'Financial Exclusion Rate by Area', 'Urban/Rural', 'Percentage', rotation=45)
```

```python
# Call the function for digital inclusion rate by area
plot_group_by(df, 'urban_or_rural', 'digital_financially_included', 'Digital Inclusion Rate by Area', 'Urban/Rural', 'Percentage')
```
